```python
from pathlib import Path
from scipy.spatial import distance
import librosa, librosa.display
import numpy as np, matplotlib.pyplot as plt, sklearn, mir_eval, IPython.display, urllib
import heapq
```

# Loading Chroma Features

```python
chromas = []
songs = []
rootdir = Path('C:/Users/imluk/Documents/4th Year/1st Sem/4450/JustFlow-Notebooks/mymusic')
for p in rootdir.glob('**/*.mp3'):
    name = str(p)
    y, sr = librosa.load(p, offset=0.0, duration=30.0)
    songs.append((y,sr,name))
    chroma = librosa.feature.chroma_stft(y, sr)
    chromas.append(chroma.flatten())
print(len(chromas))
```

# Finding the Euclidean Distance 
generating a playlist of 5 songs based on the euclidean distance from the seed song

```python
cur = 113
compare = chromas[cur]
playlist = []
print('seed song: ' + songs[cur][2])
for i in range(len(chromas)): 
    dis = -distance.euclidean(compare, chromas[i])
    heapq.heappush(playlist, (dis, songs[i]))
    if len(playlist) > 6: 
        heapq.heappop(playlist)
```

# Playing the songs

```python
combinedSongs = []
for song in playlist:
    print(song[1][2])
    for i in range(220500):
        combinedSongs.append(song[1][0][i])
IPython.display.Audio(combinedSongs, rate=songs[0][1])
```
