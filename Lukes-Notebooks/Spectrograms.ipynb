# Spectrogram
the first thing I want to do is figure out how to feed the data from a spectrogram into a model

```python
from pathlib import Path
import librosa, librosa.display
import numpy as np, scipy, matplotlib.pyplot as plt, sklearn, librosa, mir_eval, IPython.display, urllib
```

```python
# file = Path().glob('songset/Dancin.mp3')
y, sr = librosa.load('songset/Dancin.mp3')
melspectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
print(melspectrogram)
```

```python
plt.figure(figsize=(10, 4))
y_dB = librosa.power_to_db(melspectrogram, ref=np.max)
features = []
features.append(y_dB.flatten())
librosa.display.specshow(y_dB, x_axis='time', y_axis='mel', sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-frequency spectrogram')
plt.tight_layout()
plt.show()
```
